# Notebook 2:  Clustering con GMM y Algoritmos Jerárquicos

En este notebook vamos a implementar el algoritmo de clustering de Mezcla de Normales en R. El objetivo es entender como funciona este algoritmo y  aprender a utilizarlo con un ejemplo simple. Se incluye además la estrategia para definir el número de grupos k. Al final se propone un ejercicio con nuevos datos, con el fin de aprender a utilizar el algoritmo. 

```r
library(mclust)
library(cluster)
#ruta datos
#lee datos del archivo de texto como una matriz. La primera #columna se utiliza como vector de nombres de las filas
datos <-read.table("milk1.dat",row.names=1,dec=".",header=TRUE)
summary(datos)
```

```r
clus <-Mclust(datos, G=4)
summary(clus)
clus

cl <- clus$classification
centers <- t(clus$parameters$mean)
centers
cl

```

```r
BIC<-mclustBIC(datos)
plot(BIC)
summary(BIC)
```


Ejercicio 1: <br>
a) Explique como funciona el método Mclust, describa parámetros de entrada y salida.<br>
<b>R:</b><br>
Description:        Gaussian finite mixture models fitted via EM
                    algorithm for model-based clustering,
                    classification, and density estimation, including
                    Bayesian regularization, dimension reduction for
                    visualisation, and resampling-based inference.<br>

Mclust(data, G = NULL, modelNames = NULL, 
       prior = NULL, 
       control = emControl(), 
       initialization = NULL, 
       warn = mclust.options("warn"), 
       x =  NULL, 
       verbose = interactive(), …)<br>
INPUT:<br>
data: vector numerico, matriz o dataframe de observaciones<br>
G: vector de enteros especificando el numero de componentes de la mezcla<br>
modelNames: vector de strings indicando los modelos a ser ajustados en la fase EM del clustering<br>
OUTPUT:<br>
parameters: una lista de componentes:<br>
    pro(vector con proporciones de mezcla de cada componente del modelo),mean(media para cada componente), variance(lista de parametros de varianza para el modelo)<br>
z: matriz en la cual la entrada(i,k) es la probabilidad que la observacion i en los datos de prueba pertenezca a la k-ésima clase<br>
classification: clasificacion correspondiente a z<br>
Mas informacón en: https://www.rdocumentation.org/packages/mclust/versions/5.4.1/topics/Mclust 
                    

b) Interprete y comente los resultados obtenidos. <br>
<b>R:</b> El ejercicio fuerza a que se formen 4 grupos, por lo tanto busca dentro de los modelos posibles el que tenga la menor funcion de perdida con G=4, en este caso VEV, luego como de solicitan los valores medios, despliega el valor medio de cada componente entre los integrantes de cada grupo.<br>
Después de eso, se calculó directamente el BIC, generando un sumario y una representación gráfica, permitiendo observar los modelos mas convenientes y con que G logran la mayor verosimilitud (Verosimilitud es inversa a Funcion de perdida)



```r
#Análisis de Componentes Principales para graficar los grupos obtenidos
#Calcula plano de componentes principales y grafica las observaciones con el número de su grupo
res <- princomp(datos, cor=TRUE)
res2 <- predict(res,centers)

biplot(res,xlabs=cl)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
points(res2[,1:2],col=1:9,pch=19)
```

```r
library(mclust)
library(cluster)

iris<-read.csv("Iris.csv",sep=",",header=TRUE)
# names(iris)
features<-iris[2:5]
class<-iris$Species
summary(iris)
# features
```

```r
library(cluster)
#K-means
KMEANS<-kmeans(features,3,iter.max=1000,nstart=10)
# pairs(features,col=KMEANS$cluster)
# res <- princomp(features, cor=TRUE)
# res2 <- predict(res,KMEANS$centers)
# res2
# KMEANS
# biplot(res,xlabs=KMEANS$cluster)
# points(res2[,1:2],col=1:3,pch=19)
tablaK<-table(class,KMEANS$cluster)
tablaK
index1=max(tablaK[1,1:3])
index2=max(tablaK[2,1:3])
index3=max(tablaK[3,1:3])
purityK<-((index1)+(index2)+(index3))/150
purityK
```

```r
# BIC<-mclustBIC(features)
# summary(BIC)
GMM <-Mclust(features, G=3)
# summary(GMM)
# GMM

# cl <- GMM$classification
# centers <- t(GMM$parameters$mean)
# centers
# cl
tablaGMM<-table(class,GMM$classification)
tablaGMM
purityGMM<-((tablaGMM[1,1])+(tablaGMM[2,2])+(tablaGMM[3,3]))/150
purityGMM
```

```r
JERARQUICO <- hclust(dist(features),method="average")
res<-cutree(JERARQUICO,3)
tablaH<-table(class,res)
tablaH
purityH<-((tablaH[1,1])+(tablaH[2,2])+(tablaH[3,3]))/150
purityH
```

El modelo con mayor precisión en este caso es GMM, siendo bastante estable mientras que k-means varía en cada iteración, y el modelo jerérquico varia bastante dependiendo de la metodología utilizada(en este caso se utilizó AVERAGE).
